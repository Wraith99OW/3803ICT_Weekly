```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load job market
jm = pd.read_csv("job-market(2).csv")
jm = jm.dropna()
```

```python
%matplotlib inline

y_axis = jm["Classification"].value_counts().keys().tolist()
x_axis = jm["Classification"].value_counts().tolist()

fig = plt.figure(figsize=(15,10))

plt.barh(y_axis, x_axis, color="maroon")

plt.xlabel("Number of postings")
plt.ylabel("Classification")
plt.title("Jobs and number of postings")
```

```python
import matplotlib
#create dataframe for salary ranges
LSDf = jm.groupby('LowestSalary').count()
LSDf["LowestSalary"] = LSDf.index
LSDf["Label"] = LSDf["LowestSalary"]
LSDf["Count"] = LSDf["Id"]
LSDf = LSDf.drop("LowestSalary", axis=1)
LSDf.head()
```

```python
#salary ranges plot
my_circle = plt.Circle( (0,0), 0.4, color='white')
LSDf.plot(kind='pie',y='Count', legend=None, autopct='%1.2f%%', ylabel="")
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
locationDf = jm[jm['Location'].isna()==False]
locationDf = locationDf.groupby('Location').count()
locationDf['Location'] = locationDf.index
locationDf['Count'] = locationDf["Id"]
locationDf.head()
```

```python
%matplotlib inline
# Pie char for job market share of cities
from matplotlib import cm
locationDf.plot(kind='pie',y='Count', legend=None, autopct='%1.2f%%', ylabel="")

```
